```python
import pandas as pd
import requests
```

###**API URL: https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi**

####**1. Buatlah API request untuk mendapatkan data return API berupa JSON dengan menggunakan Python**

```python
url = "https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi"

response = requests.get(url)
data = response.json()

# Lihat data
print(data)
```

####**2. Ambil value dari key 'data' dalam JSON, dan transformasi menjadi sebuah DataFrame**

```python
# Mengambil nilai dari kunci 'data'
data = data['data']

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Menampilkan DataFrame dalam bentuk tabel
print(df.to_string(index=False))
```

```python
df
```

####**3. Transformasi data kolom 'isoDate' menjadi tipe data datetime**

```python
# Sebelum dirubah menjadi datetime
df.dtypes
```

#####**Tipe data kolom 'isoDate' sebelum dirubah adalah object**

```python
df['isoDate'] = pd.to_datetime(df['isoDate'])
```

```python
# Setelah dirubah menjadi datetime
df.dtypes
```

####**4. Aggregasi data berdasarkan kolom 'creator', serta urutkan berdasarkan jumlah berita terbanyak**

```python
aggregated_df = df.groupby('creator').size().reset_index(name='jumlah_berita').sort_values('jumlah_berita', ascending=False)

print(aggregated_df)
```
